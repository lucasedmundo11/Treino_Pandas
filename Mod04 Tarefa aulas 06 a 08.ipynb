####  2) Melhorando a interação com o API
Observe dois URLs de consultas diferentes, por exemplo o URL utilizado em aula, e o URL feito no exercício anterior. Compare-os e observe as diferenças.

1. Faça uma função em Python que recebe como argumento o UF da consulta e o tipo interessado (```'Estado'```ou ```Município```), e que devolve os dados da consulta no formato *DataFrame*.
2. Quantas solicitações para o Estado podem ser consultadas para Minas Gerais com *status* em 'Arquivado por decurso de prazo' estão registradas?
3. Qual é o município da Bahia com mais solicitações deferidas?
4. Salve um arquivo .csv com os dados de solicitações da Bahia, com tipo interessado = 'Estado'

````python
#1) Seu código aqui

#'https://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl?uf=RJ&tipo_interessado=Munic%C3%ADpio'
#'http://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl?uf=SP&tipo_interessado=Estado'

# Função para o retorno do DataFrame, partindo do link criado na função consulta_api
def detaframe_api(url:str):
    r = requests.get(url)
    if r.status_code == 200:
        data_json = r.json()
        return pd.DataFrame(data_json['items'])
    return False
````

````python
# Função utilizada para a criação da url da api
def consulta_api(uf:str, tp_interessado:str):
    if tp_interessado[0] in ('Mm'):
        url = 'https://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl?uf=' + uf.upper() + '&tipo_interessado=Munic%C3%ADpio'
        return detaframe_api(url)
    elif tp_interessado[0] in ('Ee'):
        url = 'https://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl?uf=' + uf.upper() + '&tipo_interessado=Estado'
        return detaframe_api(url)
    print('Favor digitar Estado ou Município para o tipo interessado.')
    return None
    
````

````python
# 2) Seu código aqui

# Realizando a consulta para o Estado de Minas Gerais 
df_mg = consulta_api('MG', 'Estado')
df_mg
````

````python
# Contando apenas o resultado desejado
df_mg[df_mg['status'] == 'Arquivado por decurso de prazo'].status.value_counts()
````

````python
# Alternativa - Contando todos os possíveis resultados
df_mg.status.value_counts()
````

````python
# 3) Seu código aqui

# Realizando a consulta para os municípios da Bahia 
df_ba = consulta_api('BA', 'Município')
df_ba
````

````python
# Contando o número de requisições por município
df_ba.loc[df_ba['status'] == 'Deferido', 'interessado'].value_counts()
````

````python
# 4) Seu código aqui

# Realizando a consulta para o Estado da Bahia 
df_ba_estado = consulta_api('BA', 'Estado')
df_ba_estado
````

````python
# Salvando o DataFrame da Bahia em um arquivo .csv
df_ba_estado.to_csv('dataframe_requisicoes_bahia.csv', index= False, sep=';', decimal=',')
````
